```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/Admin/OneDrive/Desktop/oo.csv')
df
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
# Select a few random countries to visualize
sample_countries = df['Entity'].sample(10, random_state=1).unique()
sample_df = df[df['Entity'].isin(sample_countries)]

# Create a line plot for the prevalence of undernourishment
plt.figure(figsize=(14, 8))
for country in sample_countries:
    country_data = sample_df[sample_df['Entity'] == country]
    sns.lineplot(x='Year', y='2.1.1 Prevalence of undernourishment', data=country_data, label=country)

plt.xlabel('Year')
plt.ylabel('2.1.1 Prevalence of Undernourishment (%)')
plt.title('Prevalence of Undernourishment Over Time for Selected Countries')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the average prevalence of undernourishment for each country
avg_prevalence = sample_df.groupby('Entity')['2.1.1 Prevalence of undernourishment'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(14, 8))
sns.barplot(x='Entity', y='2.1.1 Prevalence of undernourishment', data=avg_prevalence)
plt.xticks(rotation=90)

plt.xlabel('Country')
plt.ylabel('Average 2.1.1 Prevalence of Undernourishment (%)')
plt.title('Average Prevalence of Undernourishment for Selected Countries')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(x='2.1.1 Prevalence of undernourishment', y='percentage', hue='Entity', data=sample_df, palette='viridis')

plt.xlabel('2.1.1 Prevalence of Undernourishment (%)')
plt.ylabel('Percentage')
plt.title('Scatter Plot of Prevalence of Undernourishment vs. Percentage')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
avg_prevalence = df.groupby('Entity')['2.1.1 Prevalence of undernourishment'].mean().reset_index()

# Get the top 10 countries with the highest prevalence of undernourishment
top_10_countries = avg_prevalence.nlargest(10, '2.1.1 Prevalence of undernourishment')

# Create a pie chart for the top 10 countries
plt.figure(figsize=(12, 8))
plt.pie(top_10_countries['2.1.1 Prevalence of undernourishment'], labels=top_10_countries['Entity'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("viridis", 10))
plt.title('Distribution of Prevalence of Undernourishment in Top 10 Countries')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Calculate the average prevalence of undernourishment for each country
avg_prevalence = df.groupby('Entity')['2.1.1 Prevalence of undernourishment'].mean().reset_index()

# Get the top 10 countries with the highest prevalence of undernourishment
top_10_countries = avg_prevalence.nlargest(10, '2.1.1 Prevalence of undernourishment')

# Filter the original dataframe to only include the top 10 countries
top_10_df = df[df['Entity'].isin(top_10_countries['Entity'])]

# Create a line plot for the prevalence of undernourishment over time
plt.figure(figsize=(14, 8))
for country in top_10_countries['Entity']:
    country_data = top_10_df[top_10_df['Entity'] == country]
    sns.lineplot(x='Year', y='2.1.1 Prevalence of undernourishment', data=country_data, label=country)

plt.xlabel('Year')
plt.ylabel('2.1.1 Prevalence of Undernourishment (%)')
plt.title('Prevalence of Undernourishment Over Time for Top 10 Countries')
plt.legend()
plt.grid(True)
plt.show()
```
